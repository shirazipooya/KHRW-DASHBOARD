```python
import pandas as pd

xls = pd.ExcelFile('DATA.xlsx')
Data = pd.read_excel(xls, sheet_name='Data')
Data = Data.drop_duplicates()
GeoInfo = pd.read_excel(xls, sheet_name='GeoInfo')
```

```python
DATA_merged = pd.merge(Data, GeoInfo, how="outer", on=["code", "name_mahal"])
DATA_merged.sort_values(by=["code", "name_mahal"]).to_excel('DATA_merged.xlsx', index=False)
```
